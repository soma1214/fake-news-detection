```python
import numpy as np
import pandas as pd
import re
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.naive_bayes import MultinomialNB
import nltk
nltk.download('stopwords')
```

```python
df=pd.read_csv("fake_and_real_news.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
len(df['Text'])
```

```python
ps=PorterStemmer()
```

```python
##Text preprocessing
corpus=[]
for i in range(len(df['Text'])):
  news=re.sub('[^a-zA-Z]',' ',df['Text'][i])
  news=news.lower()
  news=news.split()
  news=[ps.stem(word) for word in news if word not in set(stopwords.words('english'))]
  news=' '.join(news)
  corpus.append(news)
```

```python
type(corpus)
```

```python
##create a bag of words
df['label'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['label']=le.fit_transform(df['label'])
```

```python
df
```

```python
df['label'].isnull().sum()
```

```python
y=df['label']
```

```python
X_train,X_test,y_train,y_test=train_test_split(corpus,y,test_size=0.2,random_state=0)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
len(y_train)
```

```python
len(y_test)
```

```python
###Text->vectors
vectorizer=TfidfVectorizer()
X_train=vectorizer.fit_transform(X_train).toarray()
X_test=vectorizer.transform(X_test).toarray()
```

```python
X_train
```

```python
X_test
```

```python
vectorizer.vocabulary_
```

```python
fake_news_detection=MultinomialNB()
fake_news_detection.fit(X_train,y_train)
```

```python
y_pred=fake_news_detection.predict(X_test)
```

```python
y_pred
```

```python
accuracy_score=accuracy_score(y_test,y_pred)
confusion_matrix=confusion_matrix(y_test,y_pred)
classification_report=classification_report(y_test,y_pred)
```

```python
print("Accuracy Score: ",accuracy_score)
print("Confusion Matrix: ",confusion_matrix)
print("Classification Report: ",classification_report)
```

```python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
logistic.fit(X_train,y_train)
```

```python
y_pred_logic=logistic.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Calculate accuracy score
accuracy = accuracy_score(y_test, y_pred_logic)

# Generate confusion matrix
confusion_mat = confusion_matrix(y_test, y_pred_logic)

# Generate classification report
classification_rep = classification_report(y_test, y_pred_logic)

# Print the results
print("Logistic Regression Accuracy Score:", accuracy)
print("Logistic Regression Confusion Matrix:\n", confusion_mat)
print("Logistic Regression Classification Report:\n", classification_rep)
```

```python
import pickle

# Save the vectorizer
with open('tfidf_vectorizer.pkl', 'wb') as f:
    pickle.dump(vectorizer, f)

# Save Naive Bayes classifier
with open('nb_model.pkl', 'wb') as f:
    pickle.dump(fake_news_detection, f)

# Save Logistic Regression classifier
with open('logistic_model.pkl', 'wb') as f:
    pickle.dump(logistic, f)
```

```python
import pickle

# Load the vectorizer
with open('tfidf_vectorizer.pkl', 'rb') as f:
    vectorizer = pickle.load(f)

# Load the Naive Bayes model
with open('nb_model.pkl', 'rb') as f:
    fake_news_detection = pickle.load(f)

# Alternatively, load the Logistic Regression model
with open('logistic_model.pkl', 'rb') as f:
    logistic = pickle.load(f)
```

```python
# Example new news article
new_news = "Breaking news: The government has announced new policies today."

# Preprocess the new news similar to training
import re
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')

ps = PorterStemmer()

# Clean and preprocess
news = re.sub('[^a-zA-Z]', ' ', new_news)
news = news.lower()
news = news.split()
news = [ps.stem(word) for word in news if word not in set(stopwords.words('english'))]
processed_news = ' '.join(news)
```

```python
# Transform using the same vectorizer
vectorized_news = vectorizer.transform([processed_news]).toarray()

# Predict with Naive Bayes
nb_prediction = fake_news_detection.predict(vectorized_news)

# Or predict with Logistic Regression
logistic_prediction = logistic.predict(vectorized_news)

# Output
if nb_prediction[0] == 0:
    print("Naive Bayes Prediction: Fake News")
else:
    print("Naive Bayes Prediction: Real News")

if logistic_prediction[0] == 0:
    print("Logistic Regression Prediction: Fake News")
else:
    print("Logistic Regression Prediction: Real News")
```
